# Calculate rate of growth
* To calculate growth rate: new value - old value / old value
* For D0 rate, starting from January, when we have a full month of values
* For D4 rate, starting from April, when a value first appeared in 2011

```python
(df.query("month == 'July'").D0.mean()-df.query("month == 'January'").D0.mean())/df.query("month == 'January'").D0.mean()
```

```python
(df.query("month == 'July'").D4.mean()-df.query("month == 'April'").D4.mean())/df.query("month == 'April'").D4.mean()
```

```python
# Adding an extra calculation between April and May where I see the biggest jump in D4 mean values
(df.query("month == 'May'").D4.mean()-df.query("month == 'April'").D4.mean())/df.query("month == 'April'").D4.mean()
```

```python
# Actually going to do it month to month...
(df.query("month == 'June'").D4.mean()-df.query("month == 'May'").D4.mean())/df.query("month == 'May'").D4.mean()
```

```python
(df.query("month == 'July'").D4.mean()-df.query("month == 'June'").D4.mean())/df.query("month == 'June'").D4.mean()
```

# Texas 2022 drought observations

* Rate of growth for the state under any drought category (represented by D0) was 3.3% in a 7-month period
* Rate of growth for the state under exceptional drought category (D4) was 29.6% in a 4-month period or 241.2% in a 5-month period beginning in March when D4 category first began in 2022.
* Biggest jump was March to April with 163.3%
* Then it was 56.8% April to May, -30% to June, and 18.1% to July

```python
# Creating one line of code for rate of growth
# .to_frame() to show what data looks like in a dataframe
df11 = df.groupby('month').D4.mean().to_frame().reset_index()
df11 = df11.sort_values('month')
df11 = df11[(df11['month'] > '02') & (df11['month'] <'08')]
df11 = df11.D4.pct_change().to_frame().reset_index()
```

```python
df11.rename(columns={'index':'month', 'D4': 'D411_pctchng'}, inplace=True)
df22.rename(columns={'index':'month', 'D4': 'D422_pctchng'}, inplace=True)
merged = df11.merge(df22, on='month')
merged.head()
```

```python
merged.to_csv('texas-D4-rate-change.csv')
```

```python
# Let's turn our averages into df format
df22_d4 = df2.groupby('month').D4.mean().sort_index().to_frame().reset_index()
df11_d4 = df.groupby('month').D4.mean().sort_index().to_frame().reset_index()
```

```python
# Let's try again ... to turn our averages into df format
df22_d4 = df2.groupby('month').D4.mean().sort_index().to_frame().reset_index()
df11_d4 = df.groupby('month').D4.mean().sort_index().to_frame().reset_index()
```
